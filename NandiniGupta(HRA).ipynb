```python

"""
Measuring What Matters: An Analytical Study of Key HR Metrics in Organizations

Objective of the Project:
To understand and analyze key HR metrics using sample or real organizational data,
and evaluate how these metrics support evidence-based HR decision-making in
areas such as attrition, performance, diversity, employee engagement, and
workforce planning.
"""

# Cell 1: Project Setup and Data Loading

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure pandas to display all columns
pd.set_option('display.max_columns', None)

# Load the dataset
# Assuming 'HR_Analytics.csv' is uploaded to Google Colab environment or
# mounted from Google Drive.
try:
    df = pd.read_csv('HR_Analytics.csv')
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: 'HR_Analytics.csv' not found. Please ensure the file is uploaded to your Colab environment or the path is correct.")
    # Exit or handle the error appropriately if the file is critical
    exit()

# Display the first few rows of the dataframe to get an initial look
print("\nFirst 5 rows of the dataset:")
print(df.head())

# Cell 2: Section 1 - Understanding the Organization and Data

print("--- Section 1: Understanding the Organization and Data ---")

# 1.1 What type of organization or industry is the dataset representing?
# We can infer this by looking at the column names and sample data.
print("\n1.1 Inferred Organization/Industry Type:")
print("Based on the column names (e.g., 'Age', 'Department', 'JobRole', 'MonthlyIncome', 'Attrition', 'PerformanceRating', 'TrainingTimesLastYear', 'YearsAtCompany', 'EducationField', 'MaritalStatus', etc.), this dataset appears to represent a corporate or business organization, likely in a generic industry, focusing on employee characteristics and HR-related metrics. It's a general HR dataset rather than industry-specific like manufacturing or healthcare.")

# 1.2 How many employees are there in total?
total_employees = df.shape[0]
print(f"\n1.2 Total number of employees in the dataset: {total_employees}")

# 1.3 What are the key variables (columns) in the dataset, and what do they represent?
print("\n1.3 Key Variables (Columns) and their Representation:")
for col in df.columns:
    print(f"- {col}: {df[col].dtype} - (Brief description based on typical HR datasets)")
    # For a more detailed description, one would typically manually describe each meaningful column.
    # For example:
    # - Age: Employee's age (numerical)
    # - Attrition: Whether the employee left the company (Yes/No - categorical)
    # - Department: Department the employee works in (categorical)
    # - DistanceFromHome: Distance from home to work (numerical)
    # - Education: Level of education (numerical, e.g., 1-5 for Bachelor, Master, etc.)
    # - JobRole: Employee's job role (categorical)
    # - MonthlyIncome: Employee's monthly income (numerical)
    # - PerformanceRating: Employee's last performance rating (numerical)
    # - YearsAtCompany: Total years employee has worked at the company (numerical)
    # ... and so on for all relevant columns.

# 1.4 Are there any missing or inconsistent values in the dataset that need to be addressed?
print("\n1.4 Checking for Missing Values:")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
if missing_values.sum() == 0:
    print("No missing values found in the dataset.")
else:
    print("Missing values found in the above columns. Further cleaning might be required.")

print("\nChecking for Duplicate Rows:")
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
if duplicate_rows == 0:
    print("No duplicate rows found.")
else:
    print("Duplicate rows found. Consider dropping them if they are true duplicates.")

print("\nChecking for Inconsistent Values (e.g., unique values for categorical columns):")
for column in df.select_dtypes(include='object').columns:
    print(f"Unique values for '{column}': {df[column].unique()}")
    # This helps identify inconsistencies like 'Sales' vs 'sales'
    # Or 'Yes' vs 'Y' for binary columns. In this dataset, 'Attrition' is 'Yes'/'No'.

# Cell 3: Section 2 - HR Metrics Calculation

print("\n--- Section 2: HR Metrics Calculation ---")

# 2.1 What is the overall attrition rate of the organization?
# Convert 'Attrition' column to numeric (0 for No, 1 for Yes) if it's not already
df['Attrition_Numeric'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
overall_attrition_rate = df['Attrition_Numeric'].mean() * 100
print(f"\n2.1 Overall Attrition Rate: {overall_attrition_rate:.2f}%")

# 2.2 Which department has the highest attrition rate? Why might that be the case?
attrition_by_department = df.groupby('Department')['Attrition_Numeric'].mean().sort_values(ascending=False) * 100
print(f"\n2.2 Attrition Rate by Department:\n{attrition_by_department.to_string()}")
highest_attrition_dept = attrition_by_department.index[0]
highest_attrition_rate_dept = attrition_by_department.iloc[0]
print(f"\nDepartment with the highest attrition rate: {highest_attrition_dept} ({highest_attrition_rate_dept:.2f}%)")
print("Possible reasons for high attrition in this department could include:")
print("- High workload or pressure.")
print("- Lack of career growth opportunities.")
print("- Low compensation relative to market.")
print("- Poor management or team dynamics.")
print("- Repetitive or unchallenging work.")
print("- Mismatch between employee skills and job requirements.")

# 2.3 What is the average tenure of employees across the organization?
# Assuming 'YearsAtCompany' represents tenure
average_tenure_overall = df['YearsAtCompany'].mean()
print(f"\n2.3 Average Tenure of Employees across the Organization: {average_tenure_overall:.2f} years")

# 2.4 How does average tenure vary by department or job role?
average_tenure_by_department = df.groupby('Department')['YearsAtCompany'].mean().sort_values(ascending=False)
print(f"\n2.4 Average Tenure by Department:\n{average_tenure_by_department.to_string()}")
average_tenure_by_jobrole = df.groupby('JobRole')['YearsAtCompany'].mean().sort_values(ascending=False)
print(f"\nAverage Tenure by Job Role:\n{average_tenure_by_jobrole.to_string()}")

# 2.5 What is the gender ratio in the company? Are certain departments more male/female dominant?
gender_counts = df['Gender'].value_counts()
total_gender_counts = gender_counts.sum()
male_ratio = (gender_counts.get('Male', 0) / total_gender_counts) * 100
female_ratio = (gender_counts.get('Female', 0) / total_gender_counts) * 100
print(f"\n2.5 Gender Ratio in the Company: Male {male_ratio:.2f}%, Female {female_ratio:.2f}%")

gender_by_department = df.groupby(['Department', 'Gender']).size().unstack(fill_value=0)
gender_by_department['Total'] = gender_by_department.sum(axis=1)
gender_by_department['Male_Ratio'] = (gender_by_department['Male'] / gender_by_department['Total']) * 100
gender_by_department['Female_Ratio'] = (gender_by_department['Female'] / gender_by_department['Total']) * 100
print(f"\nGender Distribution by Department:\n{gender_by_department[['Male', 'Female', 'Male_Ratio', 'Female_Ratio']].to_string()}")
print("\nAnalysis of gender dominance in departments:")
for dept in gender_by_department.index:
    male_r = gender_by_department.loc[dept, 'Male_Ratio']
    female_r = gender_by_department.loc[dept, 'Female_Ratio']
    if male_r > female_r + 10: # 10% threshold for dominance
        print(f"- {dept}: Appears Male dominant ({male_r:.2f}% Male)")
    elif female_r > male_r + 10:
        print(f"- {dept}: Appears Female dominant ({female_r:.2f}% Female)")
    else:
        print(f"- {dept}: Relatively balanced gender distribution.")

# 2.6 Which department has the highest average salary?
# Assuming 'MonthlyIncome' represents salary
average_salary_by_department = df.groupby('Department')['MonthlyIncome'].mean().sort_values(ascending=False)
print(f"\n2.6 Average Monthly Income by Department:\n{average_salary_by_department.to_string()}")
highest_salary_dept = average_salary_by_department.index[0]
highest_salary_value = average_salary_by_department.iloc[0]
print(f"\nDepartment with the highest average monthly income: {highest_salary_dept} (Avg. Income: ${highest_salary_value:.2f})")

# 2.7 How does performance rating relate to tenure or training hours?
print("\n2.7 Relationship between Performance Rating, Tenure, and Training Hours:")

# Performance vs Tenure
# Group by PerformanceRating and calculate average YearsAtCompany
performance_vs_tenure = df.groupby('PerformanceRating')['YearsAtCompany'].mean()
print(f"\nAverage YearsAtCompany by PerformanceRating:\n{performance_vs_tenure.to_string()}")
# Plotting
plt.figure(figsize=(8, 5))
sns.barplot(x=performance_vs_tenure.index, y=performance_vs_tenure.values, palette='viridis')
plt.title('Average Years at Company by Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Average Years at Company')
plt.show()


# Performance vs Training Hours (assuming TrainingTimesLastYear)
# Group by PerformanceRating and calculate average TrainingTimesLastYear
performance_vs_training = df.groupby('PerformanceRating')['TrainingTimesLastYear'].mean()
print(f"\nAverage TrainingTimesLastYear by PerformanceRating:\n{performance_vs_training.to_string()}")
# Plotting
plt.figure(figsize=(8, 5))
sns.barplot(x=performance_vs_training.index, y=performance_vs_training.values, palette='magma')
plt.title('Average Training Times Last Year by Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Average Training Times Last Year')
plt.show()


# 2.8 Is there a correlation between training hours and performance scores?
# Assuming 'TrainingTimesLastYear' as training hours and 'PerformanceRating' as performance scores
correlation_training_performance = df['TrainingTimesLastYear'].corr(df['PerformanceRating'])
print(f"\n2.8 Correlation between Training Times Last Year and Performance Rating: {correlation_training_performance:.3f}")
if correlation_training_performance > 0.5:
    print("There is a strong positive correlation, suggesting more training might lead to higher performance.")
elif correlation_training_performance > 0.1:
    print("There is a weak to moderate positive correlation.")
elif correlation_training_performance < -0.1:
    print("There is a weak to moderate negative correlation.")
else:
    print("There is a very weak or no significant linear correlation.")

# Visualizing the relationship (e.g., scatter plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TrainingTimesLastYear', y='PerformanceRating', data=df, hue='Attrition', palette='coolwarm', alpha=0.6)
plt.title('Scatter Plot: Training Times Last Year vs. Performance Rating')
plt.xlabel('Training Times Last Year')
plt.ylabel('Performance Rating')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Cell 4: Section 3 - Interpretation and Insights

print("\n--- Section 3: Interpretation and Insights ---")

# 3.1 What trends or anomalies did you observe in attrition or performance data?
print("\n3.1 Trends/Anomalies in Attrition and Performance Data:")
print("- Attrition often shows higher rates in specific departments (e.g., Sales, HR, etc.) or for employees with certain tenure ranges (e.g., short tenure due to onboarding issues, or mid-tenure due to lack of growth).")
print("- Performance ratings might cluster around certain values, indicating a potential 'bell curve' distribution or a tendency for managers to rate employees similarly.")
print("- Anomalies could include unusually high attrition in a typically stable department, or a sudden dip in performance across a group of employees.")

plt.figure(figsize=(10, 6))
sns.histplot(df[df['Attrition'] == 'Yes']['Age'], bins=20, kde=True, color='red', label='Attrition: Yes')
sns.histplot(df[df['Attrition'] == 'No']['Age'], bins=20, kde=True, color='blue', label='Attrition: No', alpha=0.6)
plt.title('Age Distribution by Attrition')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='PerformanceRating', hue='Attrition', data=df, palette='viridis')
plt.title('Performance Rating Distribution by Attrition')
plt.xlabel('Performance Rating')
plt.ylabel('Number of Employees')
plt.show()


# 3.2 Are there any patterns in employee exits (e.g., by gender, tenure, department)?
print("\n3.2 Patterns in Employee Exits:")
# Attrition by Gender
attrition_gender = df.groupby('Gender')['Attrition_Numeric'].mean() * 100
print(f"\nAttrition Rate by Gender:\n{attrition_gender.to_string()}")

# Attrition by Tenure (YearsAtCompany) - Binned
df['Tenure_Group'] = pd.cut(df['YearsAtCompany'], bins=[0, 2, 5, 10, 20, np.inf],
                            labels=['<2 years', '2-5 years', '5-10 years', '10-20 years', '>20 years'], right=False)
attrition_tenure = df.groupby('Tenure_Group')['Attrition_Numeric'].mean().sort_values(ascending=False) * 100
print(f"\nAttrition Rate by Tenure Group:\n{attrition_tenure.to_string()}")

# Attrition by Department (already calculated in Section 2)
print(f"\nAttrition Rate by Department (recap):\n{attrition_by_department.to_string()}")

print("\nObserved patterns:")
print("- Check if a particular gender has a higher attrition rate.")
print("- See if attrition is higher among new employees (e.g., <2 years) or long-term employees.")
print("- Reiterate department-specific attrition insights.")

# 3.3 Which group of employees appears to be most engaged or at risk of leaving?
print("\n3.3 Identifying Engaged/At-Risk Groups:")
print("- Employees with lower Performance Ratings and higher attrition probability are likely at risk.")
print("- Employees with short tenure (e.g., <2 years) who attrite often indicate issues with onboarding, job fit, or initial compensation.")
print("- Employees in departments with high attrition rates are naturally more at risk.")
print("- High performers with low training hours might indicate a need for more investment.")
print("- Employees who commute long distances (DistanceFromHome) might also be at higher risk of leaving.")
print("  - Attrition by DistanceFromHome (binned):")
df['Distance_Group'] = pd.cut(df['DistanceFromHome'], bins=[0, 5, 15, 30, np.inf],
                              labels=['<5 miles', '5-15 miles', '15-30 miles', '>30 miles'], right=False)
attrition_distance = df.groupby('Distance_Group')['Attrition_Numeric'].mean().sort_values(ascending=False) * 100
print(attrition_distance.to_string())

# 3.4 Based on your findings, where should HR focus its retention efforts?
print("\n3.4 Focus Areas for HR Retention Efforts:")
print("- Departments with highest attrition rates.")
print("- Employees in their initial tenure period (e.g., first 2-5 years).")
print("- Specific job roles showing higher attrition.")
print("- Addressing compensation or workload issues in high-attrition areas.")
print("- Improving employee engagement and development opportunities, especially for high performers.")

# 3.5 Is there evidence of diversity imbalance in the organization?
print("\n3.5 Evidence of Diversity Imbalance:")
print("- Gender ratio analysis by department already highlighted potential imbalances.")
print("- Further analysis could include 'EducationField', 'MaritalStatus', 'AgeGroup' distributions across departments or job roles to identify other diversity dimensions.")
print("  - Example: Distribution of EducationField by Department:")
edu_field_by_dept = df.groupby(['Department', 'EducationField']).size().unstack(fill_value=0)
print(f"\nEducation Field Distribution by Department:\n{edu_field_by_dept.to_string()}")
# Visualizing Education Field distribution
plt.figure(figsize=(12, 7))
sns.countplot(y='EducationField', hue='Department', data=df, palette='tab20')
plt.title('Distribution of Education Field by Department')
plt.xlabel('Count')
plt.ylabel('Education Field')
plt.show()


# 3.6 What insights can be drawn about career progression or promotions?
print("\n3.6 Insights about Career Progression/Promotions:")
print("- We can look at 'YearsSinceLastPromotion' and 'YearsInCurrentRole' in relation to 'JobLevel' and 'MonthlyIncome'.")
print("- Employees who stay longer in a low 'JobLevel' without promotions might be at higher risk of attrition.")
print("- High performers who are not promoted might also be at risk.")
print("  - Average YearsSinceLastPromotion by JobLevel:")
promotions_by_joblevel = df.groupby('JobLevel')['YearsSinceLastPromotion'].mean()
print(f"\nAverage YearsSinceLastPromotion by JobLevel:\n{promotions_by_joblevel.to_string()}")
plt.figure(figsize=(8, 5))
sns.barplot(x=promotions_by_joblevel.index, y=promotions_by_joblevel.values, palette='cubehelix')
plt.title('Average Years Since Last Promotion by Job Level')
plt.xlabel('Job Level')
plt.ylabel('Average Years Since Last Promotion')
plt.show()

# Cell 5: Section 4 - Strategic Recommendations (Placeholder for user input)

print("\n--- Section 4: Strategic Recommendations ---")
print("This section requires strategic thinking based on the analysis. Here are prompts for your recommendations:")

print("\n4.1 What HR initiatives would you recommend to improve retention or engagement?")
print("   - Based on departments with high attrition, suggest targeted programs (e.g., mentorship, workload management, compensation review).")
print("   - For early-tenure attrition, recommend enhanced onboarding, regular check-ins, and clear career pathing.")
print("   - For overall engagement, consider employee recognition programs, flexible work arrangements, and professional development.")

print("\n4.2 How can the company use analytics to improve workforce planning?")
print("   - Use attrition forecasts to predict future staffing needs.")
print("   - Analyze skill gaps by department/role to inform hiring and training strategies.")
print("   - Use performance data to identify high-potential employees for succession planning.")

print("\n4.3 Should the company invest more in employee training? Why or why not?")
print("   - If correlation between training and performance is positive, advocate for increased investment.")
print("   - Identify specific training needs based on performance gaps or future skill requirements.")
print("   - Highlight the long-term benefits of training on employee development and retention.")

print("\n4.4 How can HR metrics help align people strategy with business outcomes?")
print("   - By linking HR metrics (e.g., attrition, performance, engagement) to business KPIs (e.g., revenue, customer satisfaction, product quality).")
print("   - Demonstrate how reducing attrition leads to cost savings and increased productivity.")
print("   - Show how engaged employees drive innovation and better business results.")

print("\n4.5 What types of dashboards or reports would help HR leaders make better decisions?")
print("   - Attrition Dashboard: Overall rate, by department, job role, tenure, reasons for leaving (if available).")
print("   - Performance Dashboard: Distribution of ratings, average rating by department/manager, correlation with other metrics.")
print("   - Diversity Dashboard: Breakdown by gender, ethnicity, age, education across different organizational levels.")
print("   - Engagement Dashboard (if survey data available): Key engagement drivers, sentiment analysis, trends over time.")
print("   - Workforce Planning Dashboard: Headcount, hiring vs. attrition, skill inventory, succession pipeline.")

# Cell 6: Section 5 - Reflection on the Analytics Process (Placeholder for user input)

print("\n--- Section 5: Reflection on the Analytics Process ---")
print("This section is about your personal experience and learnings from the project. Here are prompts for your reflection:")

print("\n5.1 What challenges did you face in cleaning or analyzing the dataset?")
print("   - Common challenges include: missing data, inconsistent data entry (e.g., variations in categorical values), outliers, data types needing conversion, and understanding the meaning of certain columns without a data dictionary.")
print("   - Mentioned specific challenges you encountered while working with 'HR_Analytics.csv'.")

print("\n5.2 How has this project helped you understand the practical application of HR metrics?")
print("   - This project demonstrates how quantitative analysis can provide objective insights into HR issues.")
print("   - It highlights how metrics can move HR from 'gut feeling' decisions to 'evidence-based' strategies.")
print("   - It shows the importance of data quality and interpretation in deriving actionable recommendations.")

print("\n5.3 If you had access to more data (e.g., engagement surveys, exit interviews), what would you analyze next?")
print("   - Engagement Survey Data: Correlate engagement scores with performance, tenure, and attrition to identify drivers of engagement and dissatisfaction.")
print("   - Exit Interview Data: Understand specific reasons for attrition beyond generic 'Yes'/'No', which can inform targeted interventions.")
print("   - Compensation Data (more detailed): Analyze pay equity, competitiveness, and its impact on retention.")
print("   - Training Effectiveness Data: Measure the ROI of training programs by linking them to improved performance or reduced errors.")
print("   - Recruitment Data: Analyze time-to-hire, cost-per-hire, and source-of-hire effectiveness.")

print("\n5.4 How could this project be scaled using big data platforms or automation tools?")
print("   - Big Data Platforms (e.g., Apache Spark, Hadoop): For larger datasets, distributed processing, and integrating data from various HR systems (HRIS, ATS, Payroll).")
print("   - Automation Tools: Automate data extraction, cleaning, and report generation using tools like Airflow or custom Python scripts.")
print("   - Machine Learning: Implement predictive models for attrition (e.g., using Logistic Regression, Random Forest) or performance prediction, automatically identifying at-risk employees.")
print("   - Dashboards: Use BI tools like Tableau, Power BI, or even advanced Python libraries (Plotly Dash) for interactive, real-time HR dashboards.")

print("\n5.5 How would you explain the value of HR analytics to a non-HR executive?")
print("   - Focus on Business Impact: Translate HR metrics into business language (e.g., 'reducing attrition by X% saves Y dollars annually in recruitment and training costs').")
print("   - Strategic Decision Making: Explain how analytics provides data-driven insights to make better decisions about talent, leading to increased productivity, innovation, and profitability.")
print("   - Risk Mitigation: Highlight how analytics can identify and mitigate risks related to talent shortages, high turnover, or low performance before they significantly impact the business.")
print("   - Competitive Advantage: Emphasize how understanding workforce dynamics can give the company an edge in attracting, retaining, and developing top talent.")

```
